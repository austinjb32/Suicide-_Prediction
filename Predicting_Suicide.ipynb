```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file=('mexican_medical_students_mental_health_data.csv')
df=pd.read_csv(file)
df.head()
```

```python
df.describe()
```

```python
df=df.drop(columns=['id'])
```

```python
df.head()
```

```python
df['gender']=df['gender'].map(dict(m=1, f=0))
```

```python
df.head()
```

```python
df.hist(figsize=(20,14))
plt.show()
```

```python
df=df.drop(columns=['bed_time','wake_up_time','reported_sleep_hours','times_week_nap','nap_duration','weekly_study_hours'])
```

```python
x_data=df.drop(axis='columns',columns={'grades'})
y_data=df['grades']
y_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2)
x_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
num_cols = ['school_year','semester','age','gender','height','weight','phq1','phq2','phq3','phq4','phq5','phq6','phq7','phq8','phq9','total','thoughts_of_dropping_out','previous_depression_diagnosis','previous_depression_treatment','gad1','gad2','gad3','gad4','gad5','gad6','gad7','total','previous_anxiety_diagnosis','previous_anxiety_treatment','epw1','epw2','epw3','epw4','epw5','epw6','epw7','epw8','totsl_epw']
x_train[num_cols]=sc.fit_transform(x_train[num_cols])
x_test[num_cols]=sc.fit_transform(x_test[num_cols])
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.head()
```

```python
linearModel=LinearRegression()
linearModel=linearModel.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score


def cv_scores(model):
    scores=cross_val_score(model,x_train,y_train,cv=5,scoring='r2')
    print('CV Mean:',np.mean(scores))
    print('STD:',np.std(scores))
```

```python
cv_scores(linearModel)
```

```python
model= LinearRegression()
model.fit(x_test,y_test)
model.score(x_test,y_test)
```
